# Calculating Volume of Deals for Last 7 Days 

```python
sum_reliance = 0
sum_kotak = 0
sum_tata = 0
for i in range(0,7):
    sum_reliance += reliance_last_seven['Volume'][i]
    sum_kotak += kotak_last_seven['Volume'][i]
    sum_tata += tata_last_seven['Volume'][i]

import math    
sum_reliance = math.floor(sum_reliance/1000000)
sum_tata = math.floor(sum_tata/1000000)
sum_kotak = math.floor(sum_kotak/1000000)

y = [sum_reliance,sum_kotak,sum_tata]
x = ['Reliance','Kotak', 'Tata']

print("Total Volume for Reliance:",sum_reliance,"M")
print("Total Volume for Kotak:",sum_kotak,"M")
print("Total Volume for Tata:",sum_tata,"M")
```

# Visualising Volume of Deals for Last 7 Days 

```python
plt.figure(figsize=(20,10))
plt.bar(x,y,color='rgb')
plt.title("Volume for Last 7 Days",fontsize= 40)
plt.show()
```

# Insights from the above Bar Graph

## From the bar graph we can conclude that even Tata Steel had much larger volume of deals its stock prices are much less as compared to Reliance and Kotak. Thus in the closing price graph it clearly goes out of the league.
